```python
import speech_recognition as sr
import pyaudio
import googletrans
from googletrans import Translator
```

```python
r=sr.Recognizer()
print(sr.Microphone.list_microphone_names())
x=int(input('Enter the device index: '))
```

```python
mic=sr.Microphone(device_index=x)
def Take_data(s='en-in'):
    while True:
        with mic as source:
            print('Listening...')
            r.pause_threshold=1
            r.adjust_for_ambient_noise(source)
            audio=r.listen(source)
        try:
            query = r.recognize_google(audio,language=s)
            return query
        except Exception as e:
            return "Sorry, could you please repeat that"
```

```python
import googletrans
from googletrans import Translator
translator = Translator()
# lang=sr.Recognizer.supported_languages
# print(lang)
for x in googletrans.LANGUAGES.keys():
    print(x,' : ',googletrans.LANGUAGES[x])
s1=str(input('Enter the code of source language of PERSON-1: '))
s2=str(input('Enter the code of source langugae of PERSON-2: '))
print('')
print("DESTINATION LANGUAGES")
c=True
while(c):
    print('PERSON-1: (',googletrans.LANGUAGES[s1],'):')
    d1 = Take_data(s1)
    r1=translator.translate(d1,src=s1[0:2],dest=s2)
    print('Data: ',d1)
    print('Translated: ',r1.text)
    print('PERSON-2: (',googletrans.LANGUAGES[s2],'):')
    d2=Take_data(s2)
    print('Data: ',d2)
    r2=translator.translate(d2,src=s2[0:2],dest=s1)
    print('Translated: ',r2.text)
    ch=str(input('Do you still want to continue the conversation? (yes/no): '))
    if(ch.lower()=='yes'):
        c=True
    else:
        c=False
```
