```python
import speech_recognition as sr
import pyaudio
```

```python
r = sr.Recognizer()
```

```python
sr.Microphone.list_microphone_names()
```

```python
mic = sr.Microphone(device_index=0)
```

```python
def Take_data(s='en-in'):
    while True:
        with mic as source:
            print('Listening...')
            r.pause_threshold=1
            r.adjust_for_ambient_noise(source)
            audio=r.listen(source)
        try:
            query=r.recognize_google(audio,language=s)
            return query
        except Exception as e:
            print('Sorry\nCould you please repeat that')
            return "none"
```
